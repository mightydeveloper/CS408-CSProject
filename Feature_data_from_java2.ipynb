```python
import os
data_directory = "gait_feature_values2/"
files = os.listdir(data_directory)
print(files[:1])
print(len(files))
```

```python
import numpy as np
import pickle
from utils import *

save_dir = "java_data_pickled2/"
check_dir(save_dir)

#hyperparams = ['Max_threshold', 'min_stretch', 'max_stretch', 'min_step', 'max_step',
#               'smooth_nl', 'smooth_nr', 'smooth_degree', 'peak_min_distance', 'outlier_m']

y_mapping = {"sober":0 , "blue":1 , "black":2}

for file in files:
    with open(data_directory+file) as f:
        print('-------------------------------------------------')
        configuration = f.readline().strip().split()
        print(configuration)
#         for hyp, val in zip(hyperparams, configuration):
#             print(hyp,":", val)
        features = f.readlines()
        features = [line.strip().split() for line in features]
        n = len(features)
        num_features = len(features[0]) - 1 # last value is y
        
        X = np.zeros(shape=(n, num_features))
        y = np.zeros(shape=(n,))
        
        for idx, (*x, file_name) in enumerate(features):
            color, person, phone, number, shift = file_name[:-4].split('_') # remove ".txt" and split by '_'
            X[idx] = x
            y[idx] = y_mapping[color]
        
    input_data = (X, y)
    
    with open(save_dir+file[:-4]+".pkl", "wb") as g:
        pickle.dump(input_data, g)
```
